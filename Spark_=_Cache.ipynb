```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('Cache').getOrCreate()
```

```python
df = spark.read.format('csv').option('header','true').option('inferSchema','true').load('/content/customers.csv')
```

```python
rdd = spark.sparkContext.textFile('/content/customers.csv')
```

```python
rdd_filtered = rdd.filter(lambda x: 'Bangalore' in x)
```

```python
rdd_filtered.count()
```
