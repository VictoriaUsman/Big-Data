```python
data = [
    "customer_id, name, city, province, country, registration_date, is_active",
    "0,customer_0, Ian_Tristan, Victoria, Oriental_Mindoro, Philippines, 1983-05-14, True",
    "1,customer_1, Neil_Allan, San_Pedro, Laguna, Philippines, 1983-05-24, True",
    "2,customer_2, Ian_Lester, Via_Torino, Milan, Italy, 1983-05-14, True",
    "3,customer_3, Pamela, Via_Laguarda, Como, Italy, 1983-05-11, False",
    "4,customer_4, Jeffrey, San_Gelacio, Naujan, Philippines, 1983-05-10, False",
    "5,customer_5, Matis, Agoncillo, Batangas, Philippines, 1983-02-12, False",
    "6,customer_6, Kaye, San_Pedro, Laguna, Philippines, 1990-13-20, True",

]

# This is a sample data
```

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("RDD Operations").getOrCreate()
```

```python
rdd = spark.sparkContext.parallelize(data)
```

```python
mapper = rdd.map(lambda ian : ian != header)
```

```python
#filter() function
nohead_rdd = rdd.filter(lambda ian : ian != header)
```

```python
mapper = rdd.map(lambda ian : ian == header)
```

```python
# Parsing data
```

```python
parsed_rdd = nohead_rdd.map(parsing)
```

```python
def parsing(data):
  row = data.split(",")
  return (
      int(row[0]),
      row[1],
      row[2],
      row[3],
      row[4],
      row[5],
      row[6],
      row[7] ==" True"
  )
```

```python
parsed = nohead_rdd.map(parsing)
```

```python
parsed.collect()
```

```python
# Advance RDD Operation
```

Extracting field Using Map()

Extracting province and Country

```python
Prov_Country = parsed.map(lambda ian : (ian[4],ian[5]))
```

```python
Prov_Country.collect()
```

Extracting Active Customers

```python
active_rdd = parsed.filter(lambda ian : ian[7]==True)
```

```python
active_rdd.collect()
```

Showing Distinct Country

```python
distinct_country = parsed.map(lambda ian : ian[5]).distinct()
```

```python
distinct_country.collect()
```

Count customer by country

```python
customer_by_country = parsed.map(lambda ian : (ian[5],1)).reduceByKey(lambda a,b : a+b)
```

```python
customer_by_country.collect()
```

using take() to see first 5 row





```python
parsed.take(5)
```

using CountByValue()

```python
count_by_country = parsed.map(lambda ian : ian[5]).countByValue()  # This is an action call
```

```python
count_by_country
```

Combining More Operations
Cities with Active customers

```python
active_customers_and_city = parsed.filter(lambda ian : ian[7]==True).map(lambda ian:ian[3])
```

```python
counter =  active_customers_and_city.collect()
counter
```
